```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
crypto = pd.DataFrame()
```

```python
crypto['Date'] = df_BTC['Date']
```

```python
crypto['BTC_Price'] = df_BTC['Close']
```

```python
crypto['ETH_Price'] = df_ETH['Close']
```

```python
crypto['USDC_Vol'] = df_USDC['Volume']
```

```python
crypto['USDT_Vol'] = df_USDT['Volume']
```

```python
crypto['USDC_Mkt_Cap'] = df_USDC['Market Cap']
```

```python
crypto['USDT_Mkt_Cap'] = df_USDT['Market Cap']
```

```python
crypto
```

```python
crypto_valid = crypto.dropna()
```

```python
crypto_valid
```

```python
crypto_valid['Date'] = pd.to_datetime(crypto_valid['Date'])
```

```python
crypto_valid.set_index('Date')
```

```python
crypto_valid
```

```python
corrMatrix = crypto_valid.corr()
corrMatrix
```

```python
#Correlation Between USDC Market Cap vs Bitcoin Price
```

```python
crypto_valid.plot.scatter('USDC_Mkt_Cap','BTC_Price')
```

```python
#Correlation Between USDT Market Cap vs Bitcoin Price
```

```python
crypto_valid.plot.scatter('USDC_Mkt_Cap','ETH_Price')
```

```python
# Let's break the data into 4 periods with booms and bursts of bitcoin price
# Period 1: 07/01/2020 - 03/31/2021
# Period 2: 04/01/2021 - 06/30/2021
# Period 3: 07/01/2021 - 10/31/2021
# Period 4: 11/01/2021 - 01/31/2022
```

```python
Period_1.corr()
```
